```python
import pandas as pd
import datetime
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
from pytse.pytse import PyTse,SymbolData
```

```python
pytse=PyTse()
symbols=pytse.symbols_data
```

```python
# function to set Date column as datetime object
def adjust_time(df):
    df['Date'] = df['Date'].apply(lambda x : datetime.datetime.strptime(str(x),'%m/%d/%Y'))
```

```python
# here we read the data of shakhes kol

df = pd.DataFrame(index = pd.date_range(start='15/9/2008', end='27/4/2022'))
df1 = pd.read_csv('shakhes_kol_boors.csv' ,encoding_errors = 'ignore')
df1 = df1.iloc[1: , :]

stock_name = 'shakhes_kol'
df1 = df1[['Date', '<CLOSE>', '<OPENINT>']]
df1.rename(columns = {'<CLOSE>':f'{stock_name}_close', '<OPENINT>':f'{stock_name}_prev_close'},inplace=True)
adjust_time(df1)
df1.set_index(['Date'],inplace=True)
shakhes = df1
```

```python
names = {'khegostar' : "IRO1GOST0001" , 'shapna' : 'IRO1PNES0001' , 'khesapa' : 'IRO1SIPA0001', 'vbank' : 'IRO1BANK0001', 'vepars':'IRO1BPAR0001', 'fameli':'IRO1MSMI0001' , 'foolad':'IRO1FOLD0001' , 'shapna':'IRO1PNES0001', 'lepars':'IRO1PELC0001', 'khepars':'IRO1PKOD0001' }
```

```python
# then we get data of stocks
dataframes = []
for name in names.keys():
    symbol=symbols[names[name]]
    symbol_history_data=symbol.get_symbol_history()
    data = pd.DataFrame.from_dict(symbol_history_data).transpose()
    data = data[['CLOSE']]
    data.rename(columns = {'CLOSE' : name}, inplace = True)
    dataframes.append(data)
```

```python
final_data = None
final_data = dataframes[0]
for i in range(1 ,len(dataframes)):
    final_data = pd.merge(final_data, dataframes[i], left_index=True, right_index=True)

    
```

```python
import datetime
def set_index(x):
    x = str(x)
    date = datetime.datetime(year = int(x[0:4]), month = int(x[4:6]), day = int(x[6:8]))
    return date

final_data['date'] = final_data.index
final_data['date'] = final_data['date'].apply(lambda x :set_index(x))
final_data.set_index('date', inplace = True)
```

```python
final_data['week'] = final_data.index.isocalendar()['week']
final_data['year'] = final_data.index.year
final_data
```

```python
final_data = shakhes.join(final_data)
```

```python
final_data = final_data.tail(2000)
```

```python
# get weekly average
def replace_by_week_average(df):
    return df.mean(axis = 0)
    

final_data2 = final_data.groupby(['year', 'week']).apply(lambda x : replace_by_week_average(x))
final_data2 = final_data2.drop(columns = ['week', 'year'])
final_data2.reset_index(inplace = True)
```

```python
def add_weekly_return(df, column_name):
    df[f'{column_name}_prev'] = df[column_name].shift()
    df[f'{column_name}_weekly_return'] = (df[column_name] - df[f'{column_name}_prev']) / df[f'{column_name}_prev']
    df.drop(columns = [f'{column_name}_prev'], inplace = True)
    
for name in names:
    add_weekly_return(final_data2, name)

final_data2
```

```python
add_weekly_return(final_data2 , 'shakhes_kol_close')
```

```python
final_data2 = final_data2.iloc[:-1]
final_data2 = final_data2.dropna()
final_data2
```

```python
def simple_regression_without_intercept(df, y_column, x_column , year):
    linear_regression = LinearRegression(fit_intercept = False)
    df = df[df['year'] == year]
    if len(df) == 0:
        return None
    y = df[y_column].to_numpy()
    x = df[x_column].to_numpy()
    x = x.reshape(-1, 1)
    y = y.reshape(-1, 1)
    linear_regression.fit(y, x)
    return linear_regression.coef_[0]


risk_free_rate = 0.0037
betas = {}
final_data2[f'stock_market_adjusted_return'] = final_data2[f'shakhes_kol_close_weekly_return'] - risk_free_rate


for name in names:
    beta_ = []
    final_data2[f'{name}_adjusted_return'] = final_data2[f'{name}_weekly_return'] - risk_free_rate  
    for year in range(2014, 2023):
        coeff = simple_regression_without_intercept(final_data2, f'{name}_adjusted_return' , f'stock_market_adjusted_return',year)
        if coeff is not None:
            beta_.append(coeff)
    betas[name] = beta_

```

```python
import numpy as np
for name in betas:
    indexes = list(range(2014,2023))
    indexes.remove(2017)
    plt.plot(indexes, betas[name], label = name)
    plt.legend()
    plt.show()
    plt.close()
```

```python
beta_ = []
final_data2['portfolio_weekly_adjusted_return'] = 0
for name in names:
    final_data2['portfolio_weekly_adjusted_return'] += final_data2[f'{name}_adjusted_return']
final_data2['portfolio_weekly_adjusted_return'] /= len(names)
for year in range(2014, 2023):
        coeff = simple_regression_without_intercept(final_data2, f'portfolio_weekly_adjusted_return' , f'stock_market_adjusted_return',year)
        if coeff is not None:
            beta_.append(coeff)
```

```python
indexes = list(range(2014,2023))
indexes.remove(2017)
plt.plot(indexes, beta_, label = 'portfolio_beta')
plt.legend()
plt.show()
plt.close()
```

```python
# rolling regression
from statsmodels.regression.rolling import RollingOLS

def rolling_regression_without_intercept(df, y_column, x_column):
    rols = RollingOLS(df[y_column], df[x_column], window=15)
    rres = rols.fit()
    params = rres.params.copy()
    params.index = np.arange(1, params.shape[0] + 1)
    return params

for name in names:
    coeff = rolling_regression_without_intercept(final_data2,  f'{name}_adjusted_return' , f'stock_market_adjusted_return')
    plt.plot(coeff.index, coeff[f'stock_market_adjusted_return'], label = name)
    plt.legend()
    plt.show()
    plt.close()
```

```python
# as we can see in the plots above the beta is not constant in time for stocks.
```

```python
import statsmodels.api as sm


def simple_regression_with_intercept(df, y_column, x_column , year):
    df = df[df['year'] == year]
    if len(df) == 0:
        return None
    y = df[y_column]
    X = df[x_column]
    X2 = sm.add_constant(X)
    est = sm.OLS(y, X2)
    est2 = est.fit()
    print(est2.summary())
for name in names:
    simple_regression_with_intercept(final_data2, f'{name}_adjusted_return' , f'stock_market_adjusted_return',year)
```

```python
# as we can see in the tables above the t-statistic for the intercepts are not significant so that we can say there is no alpha.
```

```python
def simple_regression_without_intercept(df, y_column, x_column , years):
    linear_regression = LinearRegression(fit_intercept = False)
    df = df[df['year'].isin(years)]
    if len(df) == 0:
        return None
    y = df[y_column].to_numpy()
    x = df[x_column].to_numpy()
    x = x.reshape(-1, 1)
    y = y.reshape(-1, 1)
    linear_regression.fit(y, x)
    return linear_regression.coef_[0]


risk_free_rate = 0.0037
betas = {}

years = [2020,2021,2022]
for name in names:
    beta_ = []
    coeff = simple_regression_without_intercept(final_data2, f'{name}_adjusted_return' , f'stock_market_adjusted_return',years)
    betas[name] = coeff

market_beta = simple_regression_without_intercept(final_data2, f'portfolio_weekly_adjusted_return' , f'stock_market_adjusted_return',years)
```

```python
def get_return_for_years(df, column, years):
    df = df[df['year'].isin(years)]
    return df[column].mean()
stock_returns = {}
for name in names:
    stock_returns[name] = get_return_for_years(final_data2, f'{name}_weekly_return', years)
    
```

```python
for name in names:
    plt.scatter(betas[name], stock_returns[name], label = name)
    plt.xlabel('beta')
    plt.ylabel('return')
    x = np.linspace(-0.2,1,100)
    y = market_beta * x+risk_free_rate
    plt.plot(x, y, '-r')
plt.legend()
plt.show()
```

```python
# being below the line means that the stock is over-values and being above the line means that it is under-valued.
```

```python
# 1 portfolio class
class Stock:
    def __init__(self, name, price):
        self.name = name
        self.price = price
        self.calculate_daily_return()
    
    def calculate_daily_return(self):
        self.price['prev'] = self.price['Close'].shift()
        self.price['return'] = (self.price['prev'] - self.price['Close']) / self.price['prev']
            

class Portfolio:
    def __init__(self):
        self.stocks = []
    
    def add_stock(self,stock):
        self.stocks.append(stock)
    
    def get_portfolio_return_and_risk(self):
        returns = pd.DataFrame(index = self.stocks[0].price.index, columns = ['return'])
        returns['return'] = 0
        for stock in self.stocks:
            returns['return'] = stock.price['return']
        returns['return'] = returns['return'] / len(self.stocks)
        mean = returns['return'].mean()
        std = returns['return'].std()
        return mean, std


            
```

```python
# 2 adding stocks and computing return and risk of portfolio
portfolio = Portfolio()
means = []
stds = []

for name in names:
    price = final_data[[f'{name}']]
    price = price.rename(columns = {name:'Close'})
    stock = Stock(name, price)
    portfolio.add_stock(stock)
    mean, std = portfolio.get_portfolio_return_and_risk()
    means.append(mean)
    stds.append(std)
    
```

```python
import matplotlib.pyplot as plt
plt.plot(means, label = 'return of portfolio')
plt.legend()
```

```python
plt.plot(stds, label = 'risk of portfolio')
plt.legend()
```
